```python
import numpy as np 
import matplotlib.pyplot as plt 
import cv2
```

```python
image = cv2.imread(r"C:\Users\ASUS\Desktop\UET\Image Processing\Finding-Counting-ImageSet\Counting\cat.jpg")
plt.imshow(image)
```

```python
output_image = image.copy()
plt.imshow(output_image)
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
template = cv2.imread(r'C:\Users\ASUS\Desktop\UET\Image Processing\Finding-Counting-ImageSet\Counting\cat\ear.png', 0)
```

```python
w, h = template.shape[::-1]

result = cv2.matchTemplate(gray_image, template, cv2.TM_CCOEFF_NORMED)
threshold = 0.5  # You can adjust the threshold for better accuracy
loc = np.where(result >= threshold)

# Draw rectangles around the detected cats
for pt in zip(*loc[::-1]):
    cv2.rectangle(output_image, pt, (pt[0] + w, pt[1] + h), (0, 255, 0), 2)

plt.imshow(cv2.cvtColor(output_image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```
